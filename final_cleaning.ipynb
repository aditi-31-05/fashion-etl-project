```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('messy_fashion_dataset_v2.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.columns
```

```python
df['brand'] = df['brand'].str.strip().str.upper()
```

```python
df['brand'].head()
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df['price'] = df['price'].astype(str).str.replace("rs", "")
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df['price'].head()
```

```python
df['ratingcount'] = df['ratingcount'].astype(str)
```

```python
# remove everything except digits
df['ratingcount'] = df['ratingcount'].str.replace(r'[^0-9]', '', regex=True)
```

```python
# convert to number
df['ratingcount'] = pd.to_numeric(df['ratingcount'], errors='coerce')
```

```python
df['ratingcount'].fillna(0, inplace=True)
```

```python
df['ratingcount'] = df['ratingcount'].astype('Int64')
```

```python
df = df[df['ratingcount'] >= 0]
```

```python
df['avg_rating'] = df['avg_rating'].astype(str)
```

```python
# keep only numbers and decimal point
df['avg_rating'] = df['avg_rating'].str.replace(r'[^0-9.]', '', regex=True)
```

```python
# convert to numeric
df['avg_rating'] = pd.to_numeric(df['avg_rating'], errors='coerce')
```

```python
df['avg_rating'].fillna(df['avg_rating'].mean(), inplace=True)
```

```python
df = df[(df['avg_rating'] >= 0) & (df['avg_rating'] <= 5)]
```

```python
df['avg_rating'] = df['avg_rating'].astype(float)
```

```python
df['rating_category'] = df['avg_rating'].apply(
    lambda x: 'GOOD' if x >= 4 else ('AVERAGE' if x >= 2 else 'POOR')
)
```

```python
df['avg_rating'] = df['avg_rating'].round(1)
```

```python
df['avg_rating'] = df['avg_rating'].astype(float)
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.drop(columns=['unnamed: 0'], inplace=True, errors='ignore') # df.drop(columns=['unnamed: 0'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df['price'] = "Rs. " + df['price'].astype(str)
```

```python
df['p_id'] = df['p_id'].astype(str)
```

```python
df['p_id'] = df['p_id'].str.strip()
```

```python
df['p_id'] = df['p_id'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
```

```python
df['p_id'].replace('', pd.NA, inplace=True)
df['p_id'].fillna('UNKNOWN', inplace=True)
```

```python
df['p_id'] = df['p_id'].str.upper()
```

```python
df.drop_duplicates(subset=['p_id'], inplace=True)
```

```python
df['p_id'] = ['P' + str(i) for i in range(1, len(df)+1)]
```

```python
df['colour'] = df['colour'].astype(str)
```

```python
# remove extra spaces
df['colour'] = df['colour'].str.strip()
```

```python
# make proper case (first letter uppercase)
df['colour'] = df['colour'].str.title()
```

```python
df['colour'].replace(['', 'Nan', 'None'], pd.NA, inplace=True)
df['colour'].fillna('Unknown', inplace=True)
```

```python
df['description'] = df['description'].astype(str)

# remove extra spaces
df['description'] = df['description'].str.strip()

# remove special characters (keep basic text)
df['description'] = df['description'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

# remove multiple spaces
df['description'] = df['description'].str.replace(r'\s+', ' ', regex=True)
```

```python
df['description'].replace('', pd.NA, inplace=True)
df['description'].fillna('No description', inplace=True)
```

```python
df['description'] = df['description'].str.replace(r'\bul\b|\bli\b', '', regex=True)
```

```python
df['description'] = df['description'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
df['p_attributes'] = df['p_attributes'].astype(str)

# remove extra spaces
df['p_attributes'] = df['p_attributes'].str.strip()

# remove unwanted symbols (keep :, ,)
df['p_attributes'] = df['p_attributes'].str.replace(r'[^a-zA-Z0-9:, ]', '', regex=True)

# normalize spacing
df['p_attributes'] = df['p_attributes'].str.replace(r'\s+', ' ', regex=True)
```

```python
df['p_attributes'] = df['p_attributes'].str.lower()
```

```python
df.drop_duplicates(inplace=True)
```

```python
def clean_text(col):
    col = col.astype(str)
    col = col.str.replace(r'\bul\b|\bli\b', '', regex=True)
    col = col.str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
    col = col.str.replace(r'\s+', ' ', regex=True)
    return col.str.strip()

df['description'] = clean_text(df['description'])
df['p_attributes'] = clean_text(df['p_attributes'])
```

```python
df.to_csv("clean_fashion_dataset.csv", index=False)
```
